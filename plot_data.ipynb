```python
import os,string,random
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
import numpy as np

# Data
x_labels = ['original', '+Adaptive frame processing', '+Re-cyclic memory provisioning', '+Resource-aware collaborative computing']
latency = [1500, 781, 744, 516]  # in ms
memory = [2000, 1219, 888, 870]  # in MB
lfs=18
# Create the figure and the first axis
fig, ax1 = plt.subplots()

# Plot latency on the first y-axis as a line chart
color = 'tab:blue'
ax1.set_xlabel('Optimization Steps', fontsize=lfs)
ax1.set_ylabel('Latency (ms)', fontsize=lfs, color=color)
ax1.plot(x_labels, latency, color=color, marker='o', label='Latency (ms)')
ax1.tick_params(axis='y', labelsize=lfs, labelcolor=color)
ax1.tick_params(axis='x', rotation=-90,labelsize=lfs, labelcolor='black')

# Create the second axis
ax2 = ax1.twinx()

# Plot memory on the second y-axis as a bar chart
color = 'tab:red'
ax2.set_ylabel('Memory Footprint (MB)', fontsize=lfs, color=color)
ax2.bar(x_labels, memory, color=color, alpha=0.5, label='Memory Footprint (MB)')
ax2.tick_params(axis='y', labelsize=lfs, labelcolor=color)

# Add a title
plt.title('Latency (Line Chart) and Memory Footprint (Bar Chart) vs. Optimization Steps',fontsize=lfs)

# Add legends
lines, labels = ax1.get_legend_handles_labels()
bars, bar_labels = ax2.get_legend_handles_labels()
ax1.legend(lines + bars, labels + bar_labels, prop=FontProperties(weight='bold',size=lfs-4), loc='upper right')

hash_string = ''.join(random.choices(string.ascii_letters + string.digits, k=8))

os.makedirs('./pic',exist_ok=True)
plt.savefig(f'./pic/Case_Study_{hash_string}.pdf', format='pdf')
plt.savefig(f'./pic/Case_Study_{hash_string}.png', format='png',dpi=300)
print(f"saved at ./pic/Case_Study_{hash_string}.png")

# Display the plot
plt.show()
```
